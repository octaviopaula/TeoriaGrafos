```python
def ft(pea, pc):
    if pea == 1:
        if pc == 'i':
            ret = 2
        elif pc == 'x':
            ret = 4
        elif pc == 'v':
            ret = 5   
        else:
            ret = 100 # estado de rejeição (erro)
    elif pea == 2:
        if pc == 'i':
            ret = 3
        elif pc == 'v' or pea == 'x':
            ret = 4
        else:
            ret = 100 # estado de rejeição (erro)
    elif pea == 3:
        if pc == 'i':
            ret = 4
        else:
            ret = 100 # estado de rejeição (erro)
    elif pea == 5 :
        if pc == 'i':
            ret = 6
        else:
            ret = 100 # estado de rejeição (erro)
    elif pea == 6 :
        if pc == 'i':
            ret = 3
        else:
            ret = 100 # estado de rejeição (erro)
    elif pea == 4:
        ret = 100
        
    return ret 

ent = input('Número Romano: ')
ea = 1
for c in ent:
    ea = ft(ea, c) # função de transição

if ea in {2, 3, 4, 5, 6}: # verifica se ea (estado atual) é um estado final.
    print('\nSIM, é um número romano!')
else:
    print('\nNÃO é um número romano!')
```
