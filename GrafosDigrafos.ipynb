```python
from IPython.display import clear_output
import numpy

print('1 - Grafos')
print('2 - Digrafos')
gr = int(input('\nSelecione o codigo: ')) 
clear_output(True)
nv = int(input('Número de vértices : '))
na = int(input('\nNúmero de arestas: ')) 

mat_adj = [[0] * nv for i in range(nv)]

if gr == 1:
  for a in range(na): 
    clear_output(True)
    print('Selecione o grafo :')
    print(f'Aresta {a + 1}:')
    vs = int(input('\nVértice de origem: ')) 
    vc = int(input('\nVértice de destino: ')) 
    mat_adj[vs-1][vc-1] = 1
    mat_adj[vc-1][vs-1] = 1
    clear_output(True)
else:
  for a in range(na): 
    clear_output(True)
    print('Selecione o digrafo :')
    print(f'Aresta {a + 1}:')
    vs = int(input('\nVértice de origem: '))
    vc = int(input('\nVértice de destino: ')) 
    mat_adj[vs][vc] = 1
    clear_output(True)

print('A matriz de adjacências é :')
for line in mat_adj:
  print(line)
```
