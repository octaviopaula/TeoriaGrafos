# Teoria de Grafos

# Matriz Adjacencias Grafos e Digrafos

```python
from IPython.display import clear_output

print('Opções:\n\nG - Grafo\nD - Dígrafo\n')
gd = input('Escolha: ').upper()
qv = int(input('\nQtde de vértices: '))
qa = int(input('\nQtde de arestas: '))

mat = []
for lin in range(qv): # cria a matriz com zeros em todas as posições
    mat.append([0] * qv)
for a in range(qa): # realiza a leitura das arestas
    clear_output(True) # limpa a área de saída de dados
    print(f'\nAresta {a + 1}:')
    vo = int(input('\nOrigem: '))
    vd = int(input('\nDestino: '))
    mat[vo - 1][vd - 1] = 1
    if gd == 'G': # se for um grafo insere o valor 1 na posição espelhada da matriz
        mat[vd - 1][vo - 1] = 1

clear_output(True) # limpa a área de saída de dados
print('\n\nMatriz de adjacências e o grau de cada vértice (para grafos):\n')
print('    ', end='') 
for i in range(qv): # mostra as colunas da Matriz de Adjacências
    print(f'{i + 1} ', end='')
if gd == 'G':
    print(' Grau', end='')
print('\n   ' + '--' * qv, end='')

for vo in range(qv): # percorre os vértices de origem (linha da matriz)
    print(f'\n{vo + 1} | ', end='')
    grau = 0
    for vd in range(qv): # percorre os vértices de destino (coluna da matriz)
        print(f'{mat[vo][vd]} ', end='')
        if gd == 'G' and mat[vo][vd] == 1: # se for grafo, para cada valor igual a 1, incrementa o grau
            grau += 1
            if vo == vd: # verifica se é um laço (aumenta mais 1 no grau do vértice)
                grau += 1
    if gd == 'G': # se for um grafo, mostra o grau
        print(f'  {grau}', end='')

```

# Listas de Adjacências

```python
from IPython.display import clear_output

nv = int(input('Número de vértices: '))
na = int(input('\nNúmero de arestas: '))

lis_adj = [] # lista que contém as listas de adjacências
lis_adj.append([0] * nv) # cria uma lista de vértices marcados (apenas para as buscas)
for i in range(nv): # cria as listas de adjacências dos vértices (saída)
    lis_adj.append([])

for a in range(na): # leitura das arestas
    clear_output(True)
    print(f'Aresta {a + 1}:')
    vs = int(input('\nVértice de saída: ')) # vértice saída
    vc = int(input('\nVértice de chegada: ')) # vértice de chegada
    lis_adj[vs].append(vc) # insere chegada na saída
    if vs != vc: # se não for um laço (vs igual a vc)
        lis_adj[vc].append(vs) # insere saída na chegada

clear_output(True)
print('Listas de Adjacências:\n')
for lis in lis_adj:
    print(lis)
```

# Busca em Profundidade

```python
from time import sleep

def bprof(lista, vi):
    lista[0][vi - 1] = 1  # marcar o vértice
    print(f'Marcar vértice {vi}')
    sleep(0.5)
    for v in lista[vi]: # percorre a lista do vértice vi (v é o vértice atual da lista)
        if lista[0][v - 1] == 0: # verifica se já foi marcado (0 - não e 1 - sim)
            bprof(lista, v) # busca novamente pelo vértice v (recursão)
        else:
            print(f'Vértice {v} já marcado.')
            sleep(0.5)

def busca_prof(lista, vi):
    lista[0] = [0] * nv # zerar todas as marcações de vértices
    bprof(lista, vi)

vini = int(input('Vértice inicial: '))
print(f'\nBusca em profundidade com início pelo vértice {vini}:\n')
busca_prof(lis_adj, vini) # lis_adj foi criada na célula anterior
```

# Busca em Largura

```python
from time import sleep

def busca_larg(lista, vi):
    lista[0] = [0] * nv # zerar todas as marcações de vértices
    filab = [vi] # cria a fila de próximas buscas com o vértice inicial
    lista[0][vi - 1] = 1  # marca o vértice inicial da busca
    print(f'Marcar vértice {vi}.\t\tFila: {filab}')
    sleep(0.5)
    while filab: # enquanto tiver valores na fila de próximas buscas
        vb = filab.pop(0) # retira da fila o próximo vértice da busca (vb)
        for v in lista[vb]: # percorre a lista do vértice vb (v é o vértice atual da lista)
            if lista[0][v - 1] == 0: # verifica se já foi marcado (0 - não e 1 - sim)
                lista[0][v - 1] = 1  # marcar o vértice v da busca
                filab.append(v) # insere o vértice na fila de próximas buscas
                print(f'Marcar vértice {v}.\t\tFila: {filab}')
            else:
                print(f'Vértice {v} já marcado.\t\tFila: {filab}')
            sleep(0.5)

vini = int(input('Vértice inicial: '))
print(f'\nBusca em largura com início pelo vértice {vini}:\n')
busca_larg(lis_adj, vini) # lis_adj foi criada na célula anterior
```

# Algoritmo Dijkstra

# Listas de adjacências com dados do grafo

```python
grafo = input('Arestas do grafo (origem destino custo, origem destino custo, ...): ')
grafo = [[int(v) for v in ar.split()] for ar in grafo.split(',')]
nv = max([v for ar in grafo for v in ar[:2]]) # calcula o maior vértice (número de vértices)
lis_adj = [[] for _ in range(nv)]
for ar in grafo:
    ori, des, custo = ar[0], ar[1], ar[2] # dados de uma aresta
    lis_adj[ori - 1].append([des, custo]) # insere chegada na saída
    if ori != des: # se não for um laço (vs igual a vc)
        lis_adj[des - 1].append([ori, custo]) # insere saída na chegada
print('\nListas de Adjacências:\n')
for i, lis in enumerate(lis_adj):
    print(f'{i + 1} : {lis}')
```

# Algoritmo Dijkstra

```python
from IPython.display import clear_output

def mostrar_dados(pda, pab, pvef):
    clear_output(True)
    print('\nTabela de dados Dijkstra:\n')
    for i, lin in enumerate(pda):
        print(i + 1, lin)
    print(f'\nAbertos: {pab}   Em fechamento: {pvef}')
    input('\nPressione ENTER para continuar...')   

def dijkstra(plis, pvi):
    dados = [[float('inf'), ''] for i in range(len(plis))] # tabela de dados
    dados[pvi - 1][0], dados[pvi - 1][1] = 0, '-' # custo zero para o vértice inicial
    abertos = [i + 1 for i in range(len(plis))] # lista de vértices abertos
    mostrar_dados(dados, abertos, pvi)
    while abertos: # enquanto tiver vértices abertos
        # calcular vértice com menor custo
        mc, vef = dados[abertos[0] - 1][0], abertos[0] # menor custo e vértice com menor custo (vértice em fechamento)
        for v in abertos:
            if dados[v - 1][0] < mc:
                mc = dados[v - 1][0]
                vef = v # armazena o vértice de menor custo (vértice em fechamento)
        abertos.remove(vef) # remove o vértice com menor custo dos vértices abertos
        cvef = dados[vef - 1][0] # custo do vértice em fechamento
        for lvc in plis[vef - 1]: # percorre a lista do vértice em fechamento (vértice e custo)
            v, c = lvc[0], lvc[1] # vértice e custo da lista em fechamento
            if v in abertos and cvef + c < dados[v - 1][0]: # verifica se o custo até o vértice atual é menor
                dados[v - 1][0] = cvef + c # atualiza o custo até v
                dados[v - 1][1] = vef # atualiza o vértice anterior (vef)
            mostrar_dados(dados, abertos, vef)

vi = int(input('\nVértice inicial (Dijkstra): '))
dijkstra(lis_adj, vi)
print('\nFim...')
```
